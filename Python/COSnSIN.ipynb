```python
import matplotlib.pyplot as plt
import numpy   as np
r = list(range(0,361,15))
x = [i*(np.pi/180) for i in r]
plt.plot(r,np.sin(x))
plt.plot(r,np.cos(x))
plt.title("BASIC TRIGNOMETRY")
plt.xlabel("ANGLES")
plt.ylabel("VALUES")

plt.show()
```
